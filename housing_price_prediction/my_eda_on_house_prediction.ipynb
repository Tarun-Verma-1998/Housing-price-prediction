# Scatter Plots of most correlated fatures

```python
fig = plt.figure(figsize=(14,10))

plt.subplot(3,2,1)
overall_scatterplot = pd.concat([train['OverallQual'],train['SalePrice']],axis=1)
sns.regplot(x="OverallQual",y='SalePrice',data=overall_scatterplot,scatter=True,fit_reg=True)

plt.subplot(3,2,2)
total_bsmt_sf_scatterplot = pd.concat([train['TotalBsmtSF'],train['SalePrice']],axis=1)
sns.regplot(x="TotalBsmtSF",y='SalePrice',data=total_bsmt_sf_scatterplot,scatter=True,fit_reg=True)

plt.subplot(3,2,3)
GrLivArea_scatter_plot = pd.concat([train['SalePrice'],train['GrLivArea']],axis = 1)
sns.regplot(x='GrLivArea',y = 'SalePrice',data = GrLivArea_scatter_plot,scatter= True, fit_reg=True)

plt.subplot(3,2,4)
GarageArea_scatter_plot = pd.concat([train['SalePrice'],train['GarageArea']],axis = 1)
sns.regplot(x='GarageArea',y = 'SalePrice',data = GarageArea_scatter_plot,scatter= True, fit_reg=True)

plt.subplot(3,2,5)
FullBath_scatter_plot = pd.concat([train['SalePrice'],train['FullBath']],axis = 1)
sns.regplot(x='FullBath',y = 'SalePrice',data = FullBath_scatter_plot,scatter= True, fit_reg=True)

plt.subplot(3,2,6)
YearBuilt_scatter_plot = pd.concat([train['SalePrice'],train['YearBuilt']],axis = 1)
sns.regplot(x='YearBuilt',y = 'SalePrice',data = YearBuilt_scatter_plot,scatter= True, fit_reg=True)


YearRemodAdd_scatter_plot = pd.concat([train['SalePrice'],train['YearRemodAdd']],axis = 1)
YearRemodAdd_scatter_plot.plot.scatter('YearRemodAdd','SalePrice')
```

```python
missing_values = numeric_features.isnull().sum(axis=0).reset_index()
missing_values.columns = ['column_name','missing_count']
missing_values = missing_values.loc[missing_values['missing_count']>0]
missing_values

ind = np.arange(missing_values.shape[0])
width = 0.2
fig,ax = plt.subplots(figsize=(12,3))

ax.barh(ind,missing_values.missing_count.values,color='b')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.column_name.values,rotation='horizontal')

plt.show()
```

```python
missing_values
```

```python
missing_values = categorical_features.isnull().sum(axis=0).reset_index()
missing_values.columns = ['column_name','missing_count']
missing_values = missing_values.loc[missing_values['missing_count']>0]
missing_values

ind = np.arange(missing_values.shape[0])
width = 0.2
fig,ax = plt.subplots(figsize=(12,18))

ax.barh(ind,missing_values.missing_count.values,color='red')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.column_name.values,rotation='horizontal')

plt.show()
```
